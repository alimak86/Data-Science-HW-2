```python
import numpy as data
```

```python
import pandas as pd
```

```python
list = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96",match="Коефіцієнт народжуваності у регіонах України у 1950—2014 роках") ## html is from the link 
```

```python
table = list[0] ## get a DataFrame object
```

```python
table.head()
```

```python
table.shape
```

```python
table.replace("-",data.nan)
```

```python
table.dtypes # all columns types are numeric except column = 0
```

```python
width=len(table.columns)
```

```python
pd.to_numeric(table.iloc[:,0]) ## can not cast to numeric
```

```python
table.isnull().sum() #calculate nan in each column
```

```python
table = table.drop(len(table.index)-1,axis=0) ## remove last row
```

```python
table
```

```python
means = table.iloc[:,1:width-1].mean() ## calculate means
```

```python
table = table.fillna(means) ## insert means into NaN
```

```python
ukraine_2014 = means.iloc[len(means)-1]
```

```python
ukraine_2014 ## average birht in 2014
```

```python
regions_high_birth_2014 = table[table.iloc[:,width-1]>ukraine_2014].iloc[:,0] 
```

```python
regions_high_birth_2014 ## display regions with birht > average in 2014
```

```python
maximums = table.max()
```

```python
maximums
```

```python
max_ukraine_2014 = maximums.iloc[len(maximums)-1]
```

```python
max_ukraine_2014
```

```python
regions_maximum_birth_2014 = table[table.iloc[:,width-1] == max_ukraine_2014].iloc[:,0] 
```

```python
regions_maximum_birth_2014 ## region with maximum birth
```

```python
df = pd.DataFrame({"Regions":bar_plot_x, "Birth, humans":bar_plot_y})
```
